# 레모네이드 판매 예측
*  https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/lemonade.csv



```python
#라이브러리 사용
import tensorflow as tf
import pandas as pd
```

```python
#데이터 준비
파일경로 = 'https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/lemonade.csv'
데이터 = pd.read_csv(파일경로)
데이터.head()
```

```python
#종속변수, 독립변수
독립 = 데이터[['온도']]
종속 = 데이터[['판매량']]
print(독립.shape, 종속.shape)
```

```python
#모델을 만듭니다
X = tf.keras.layers.Input(shape=[1]) #독립변수 칼럼의 개수가 하나이기 때문에 그거에 맞춰서 1이라고 작성
Y = tf.keras.layers.Dense(1)(X) #종속변수 칼럼의 개수가 1니까 1이라고 작성
model = tf.keras.models.Model(X, Y)
model.compile(loss='mse') #model.compile() => 괄호 안은 모델이 학습할 방법 정해줌
```

```python
#모델을 학습시킵니다
model.fit(독립, 종속, epochs=10) #(독립, 종속, epochs=10000, verbose=0) #verbose=0는 학습결과 출력을 꺼줌
```

```python
#모델을 이용합니다
model.predict(독립)
```
